Assignment 4:Bus ticket Discount

```python
Bus_ticket_price=500
Customer_age=int(input("Enter  your  age:"))
if Customer_age<5:
    print("FREE FOR YOU,SAFE TRAVEL")
elif 5<Customer_age<18:
    print("50% discount price is :",Bus_ticket_price-(Bus_ticket_price*50/100),"tk")
elif Customer_age<=60:
    print("30% discount price is :",Bus_ticket_price-(Bus_ticket_price*30/100),"tk")
else:
    print("full price")
```

Assignment 8:Leap year checker

```python
year=int(input("Enter  the  year :"))
if year%400==0:
    print("The year is leap year")
elif year%4==0 and year%100!=0:
    print("THe year is  leap year")
else :
    print("THE year is not leap year")

```
