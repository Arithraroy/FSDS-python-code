# Assignment:1 Email Validator (String Filtering + Loops)
üíº Real-world use: Used in signup forms or data cleaning tasks.
Task:
Check if a given email is valid ‚Äî must contain ‚Äú@‚Äù and end with ‚Äú.com‚Äù or ‚Äú.org‚Äù.

```python
def validate_email(email):
    """Checks if an email is valid (must contain '@' and end with .com or .org)."""
    email = email.lower()
    
    if "@" not in email:
        return "Invalid Email (Missing @)"
    
    if email.endswith(".com") or email.endswith(".org"):
        return "Valid Email"
    else:
        return "Invalid Email (Must end with .com or .org)"


print(f"hello@example.com: {validate_email('hello@example.com')}")
print(f"user@mail.net: {validate_email('user@mail.net')}")
```

#Assingment 3. Password Strength Checker
üíº Real-world use: Used in user authentication systems.
Task:
Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one
special character.

```python
def check_password_strength(password):
    """Checks if a password is strong (requires upper, lower, digit, and special char)."""
    
    checks = {
        'has_upper': any(c.isupper() for c in password),
        'has_lower': any(c.islower() for c in password),
        'has_digit': any(c.isdigit() for c in password),
        # Special character is anything that is NOT alpha or digit
        'has_special': any(not c.isalnum() for c in password) 
    }
    
    if all(checks.values()):
        return "Strong Password"
    else:
        missing = [key.replace('has_', '') for key, value in checks.items() if not value]
        return f"Weak Password (Missing: {', '.join(missing)})"

# Example Usage:
print(f"'Hello@123': {check_password_strength('Hello@123')}")
print(f"'weakpass': {check_password_strength('weakpass')}")
```

#Assignment:4Username Generator
üíº Real-world use: Automatically create usernames for users.
Task:
Take user‚Äôs full name and generate a short username.
Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).

```python
import random

def generate_username(full_name):
    """Generates a username: first 3 letters of first name + last 3 of last name + 2 random digits."""
    names = full_name.split()
    if len(names) < 2:
        return "Error: Provide both first and last name."
        
    first_name = names[0].lower()
    last_name = names[-1].lower()
    
    first_part = first_name[:3]
    last_part = last_name[-3:]
    
    # Generate a random number (00-99) and zero-pad it
    random_num = str(random.randint(0, 99)).zfill(2)
    
    return f"{first_part}{last_part}{random_num}"

# Example Usage:
print(f"Mamun Malitha -> {generate_username('Mamun Malitha')}")
```

#Assignment :5 SMS Word Counter
üíº Real-world use: Used in messaging apps or social media post limit check.
Task:
Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).

```python
def count_sms_stats(message):
    """Counts characters and words, and checks against the 160-character SMS limit."""
    char_count = len(message)
    word_count = len(message.split())
    
    status = "Within Limit" if char_count <= 160 else "Exceeds Limit"
    
    return f"Words = {word_count}, Characters = {char_count}, Status = {status}"

# Example Usage:
input_msg = "This is a demo message."
print(f"'{input_msg}' -> {count_sms_stats(input_msg)}")
```

#Assignment 6:Remove Duplicates from Contact List
üíº Real-world use: Data cleaning in CRM or user databases.
Task:
Given a list of names (comma separated), remove duplicates.

```python
def remove_duplicate_contacts(contact_string):
    """Removes duplicate names from a comma-separated list."""
    # Split, strip whitespace, and convert to a set to automatically remove duplicates
    names = [name.strip() for name in contact_string.split(',')]
    unique_names = sorted(list(set(names)))
    
    # Join back into a string
    return ", ".join(unique_names)

# Example Usage:
input_list = "Rahim, Karim, Rahim, Sultana"
print(f"Input: '{input_list}' -> Output: '{remove_duplicate_contacts(input_list)}'")
```

#Assignment :7. Detect Suspicious Words in Text (Moderation)
üíº Real-world use: Used in chat moderation or email filtering.
Task:
If any ‚Äúbanned‚Äù words like ['spam', 'scam', 'fake'] are found in a text, print ‚ÄúWarning‚Äù.

```python
def detect_suspicious_words(text):
    """Checks if any banned words are present in the text."""
    banned_words = ['spam', 'scam', 'fake', 'phishing']
    text_lower = text.lower()
    
    # Check if any banned word is found in the text
    if any(word in text_lower for word in banned_words):
        return "Warning: Suspicious content detected"
    else:
        return "Content appears safe"

# Example Usage:
print(f"'This is a scam offer' -> {detect_suspicious_words('This is a scam offer')}")
```

#Assignment:9. Count Hashtags and Mentions (Social Media Analytics)
üíº Real-world use: Used in analyzing social media posts.
Task:
Count how many hashtags (#) and mentions (@) appear in a post.

```python
def count_social_elements(post_text):
    """Counts the number of hashtags (#) and mentions (@) in a post."""
    
    counts = {'Hashtags': 0, 'Mentions': 0}
    
    for char in post_text:
        if char == '#':
            counts['Hashtags'] += 1
        elif char == '@':
            counts['Mentions'] += 1
            
    return f"Hashtags = {counts['Hashtags']}, Mentions = {counts['Mentions']}"

# Example Usage:
input_post = "Check out #AIWorkshop and follow @inceptionbd, and another tag #tech"
print(f"'{input_post}' -> {count_social_elements(input_post)}")
```

#Assignment:10 10. Word Reversal for Data Encryption
üíº Real-world use: Simple text obfuscation or playful encoding.
Task:
Reverse each word in a sentence while keeping word order the same.

```python
def reverse_words_in_sentence(sentence):
    """Reverses each word in a sentence while preserving word order."""
    
    # Split sentence into words
    words = sentence.split()
    
    # Reverse each word using slicing [::-1]
    reversed_words = [word[::-1] for word in words]
    
    # Join the reversed words back into a sentence
    return " ".join(reversed_words)

# Example Usage:
input_sentence = "AI is fun"
print(f"'{input_sentence}' -> '{reverse_words_in_sentence(input_sentence)}'")
```
