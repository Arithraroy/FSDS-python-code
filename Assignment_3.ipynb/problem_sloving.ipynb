Menu Price Lookup
Real-world use: Restaurant ordering apps.
Task:
Write a function get_price(menu, item) that:
 receives a dictionary menu and an item name
 returns the price if exists
 otherwise returns "Item not found"

```python
def analyze_temperatures(temp_list):
    high = []
    normal = []
    for temp in temp_list:
        if temp > 38:
            high.append(temp)
        else:
            normal.append(temp)
    return high, normal


# --- User Input ---
n = int(input("How many temperatures? "))
temps = []

for i in range(n):
    temp = float(input(f"Enter temperature {i+1}: "))
    temps.append(temp)

high_temps, normal_temps = analyze_temperatures(temps)
print("High:", high_temps)
print("Normal:", normal_temps)
```
